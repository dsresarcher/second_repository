# Data preparation



We have to normalize the variables to express them in the same range of values. In other words, normalization means 
adjusting values measured on different scales to a common scale.

```python
# Data Normalization

from sklearn.preprocessing import normalize
scaled = normalize(df)
scaled = pd.DataFrame(scaled)
scaled.head()
```

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
sse = []
for i in range(1, 5):
    kmeans = KMeans(n_clusters = i, init = 'random',random_state = 42)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,5),sse)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('clustering score')
plt.show()
```

```python
sse
```

```python
# Fit the model and predict
kmeans = KMeans(n_clusters = 2, random_state = 42)

kmeans.fit(df)
```

```python
pred = kmeans.predict(df)
pred
```

```python
kmeans.cluster_centers_
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled, pred)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/shpoo/Desktop/Data_set_for_ML/winequality_red.csv')
df
```

# Data Prepration

```python
from sklearn.preprocessing import normalize
scaled = normalize(df)
scaled = pd.DataFrame(scaled)
scaled.head()
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(10, 7))

dendrogram = sch.dendrogram(sch.linkage(scaled, method ='ward'))

plt.title('Dendrogram')
plt.xlabel('Quality of wine')
plt.ylabel('apply all features')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 2, affinity = 'euclidean', linkage = 'ward')
#cluster.fit_predict(datasubset)
pred= cluster.fit_predict(scaled)
pred
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(scaled, pred)
```

```python
outliers = db[db['Cluster']==-1]
fig2, (axes) = plt.subplots(1,2,figsize = (12,5))
sns.scatterplot('fixed acidity','quality',data = db[db['Cluster']!=-1],hue = 'Cluster', ax = axes[0],
                palette = 'Set2', legend = 'full', s = 200)
sns.scatterplot('alcohol','quality',data = db[db['Cluster']!=1],hue = 'Cluster', palette='Set2', ax=axes[1],
                legend='full',s = 200)

axes[0].scatter(outliers['fixed acidity'], outliers['quality'], s=10, label = 'outliers',c = 'k')

axes[1].scatter(outliers['alcohol'],outliers['quality'], s = 10, label = 'outliers', c = 'k')

axes[0].legend()
axes[1].legend()

plt.setp(axes[0].get_legend().get_texts(), fontsize = '12')
plt.setp(axes[1].get_legend().get_texts(), fontsize = '12')

plt.show()
```
